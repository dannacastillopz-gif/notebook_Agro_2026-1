## ¿Qué es Python?

Python es un lenguaje de programación de **alto nivel**, lo que en términos humanos significa que su sintaxis es muy parecida al inglés. Es famoso por ser "legible", lo que permite a ingenieros (que no necesariamente son programadores de oficio) escribir soluciones potentes sin romperse la cabeza con códigos complejos.

## ¿Por qué es vital en Ingeniería Agroindustrial?

En la agroindustria, manejamos variables biológicas, físicas y químicas que generan muchísimos datos. Python ayuda a conectar el campo con la eficiencia industrial mediante:

1. **Análisis de Datos:** Procesar historiales de cosecha, clima o niveles de inventario.
2. **Automatización:** Controlar sensores de humedad, temperatura en silos o bandas transportadoras.
3. **Optimización:** Calcular las mezclas exactas de ingredientes para alimentos balanceados o fertilizantes al menor costo.
4. **Visión Artificial:** Identificar frutas maduras o detectar plagas mediante fotos.

## Ejemplos de Código (Sencillos y Prácticos)

Aquí tienes tres ejemplos básicos que podrías empezar a usar hoy mismo:

### 1. Estimación de Rendimiento de Cosecha

Este código calcula cuántas toneladas se esperan recoger en un lote basándose en el área y un rendimiento promedio.

```
# Datos de entrada
area_hectareas = 5.5
rendimiento_por_ha = 12.3  # Toneladas por hectárea

# Cálculo
total_estimado = area_hectareas * rendimiento_por_ha

print(f"La cosecha estimada para {area_hectareas} ha es de {total_estimado} toneladas.")
```

### 2. Monitor de Temperatura en Almacenamiento

Un script simple que evalúa si la temperatura de un silo de granos es segura para evitar la proliferación de hongos.

```
temperatura_actual = 28.5  # Grados Celsius

if temperatura_actual > 25.0:
    print("¡ALERTA! Temperatura alta. Activar sistemas de ventilación.")
else:
    print("Temperatura óptima. El grano está seguro.")
```

### 3. Cálculo de Humedad en Base Seca

Fundamental en procesos de secado de granos o frutas para control de calidad.

```
peso_humedo = 150.0  # gramos
peso_seco = 120.0   # gramos

# Fórmula de humedad (Base Seca)
humedad_bs = ((peso_humedo - peso_seco) / peso_seco) * 100

print(f"El porcentaje de humedad en base seca es: {humedad_bs}%")
```
